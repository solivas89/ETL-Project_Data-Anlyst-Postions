```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
# Put data in pd Dataframe
csv_file = "../ETL-Project_Data-Anlyst-Postions/alldata.csv"
major_cities_df = pd.read_csv(csv_file)
# Take a peek at the first few rows of Dataframe
major_cities_df.head()
```

```python
# Take a peak at the last few rows of Dataframe to consistency
major_cities_df.tail()
```

```python
# Check the columns of data

major_cities_df.columns
```

```python
# Find out how many rows and columns are in data
major_cities_df.shape
```

```python
# Clean 'location column to get rid of area codes. Split the column column into three and rename
new = major_cities_df["location"].str.split(" ", n = 2, expand = True)

#  Rename first column
major_cities_df["loca"]= new[0]

# Rename secopnd column
major_cities_df["state"]= new[1]

# Rename third column
major_cities_df["area code"]= new[2]

# Drop original 'location' coulmn
major_cities_df.drop(columns =["location"], inplace = True)
major_cities_df
```

```python
# Take a peek new columns
major_cities_df.columns
```

```python
# Concatenate columns 'loca and 'state and rename the combined column 'location'
major_cities_df['location'] = major_cities_df['loca']+" "+ major_cities_df['state']
major_cities_df['location']
```

```python
#  Drop coumns 'loca', 'state' and 'area code'
major_cities_df=major_cities_df.drop(['description','loca', 'state', 'area code'], axis=1)
major_cities_df
```

```python
#  Check for null cells

major_cities_df.isnull().values.any()
```

```python
# Replace null values with unknown

# major_cities_df.replace(to_replace = "NaN", value = "Unknown", inplace = True)
# la_ny_df.replace(to_replace = "-1", value = "Unknown", inplace = True)
cleaned_cities_df = major_cities_df.fillna("Unknown")
cleaned_cities_df.head()
```

```python
cleaned_major_cities_df = cleaned_cities_df[['position', 'company', 'reviews','location']].copy()
cleaned_major_cities_df.head()
```
